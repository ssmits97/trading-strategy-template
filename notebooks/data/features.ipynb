# S&P500 Features

Let's see how our features turned out

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('../../data/features/sp500_basics.csv', parse_dates=['Date'])
df = df.set_index('Date')
df
```

```python
df[['returns', 'logs', 'Long Interest Rate']].plot()
```

```python
df['Consumer Price Index'].plot()
```

```python
mu, std = stats.norm.fit(df['returns'].dropna())

plt.hist(df['returns'], density=True)

x = np.linspace(df['returns'].min(), df['returns'].max(), 1000)
y = stats.norm.pdf(x, mu, std)
plt.plot(x, y)

plt.show()
```

Our features do not look weird.
However, we see that our returns are not normally distributed.
We cannot use the normal distribution for our returns
